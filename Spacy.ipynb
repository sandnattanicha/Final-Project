```python
from tqdm import tqdm_notebook
```

```python
import pandas as pd
df = pd.read_csv('data/text.csv')
```

```python
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm", )
doc = nlp(df['text'][1])
```

```python
import IPython
IPython.display.HTML(displacy.render(doc, style="ent"))
```

```python
def replace_by_type(texts):
    new_texts = []
    for doc in tqdm_notebook(nlp.pipe(texts, disable=["tagger", "parser"], n_threads=4), total=len(texts)):
        new_txt = doc.text
        for ent in doc.ents:
            new_txt = new_txt.replace(ent.text, ent.label_)
        new_texts.append(new_txt)
    return new_texts
```

```python
new_text = []
for i in df['text']:
    train_texts_type = replace_by_type(i)
    new_text.append(''.join(train_texts_type))
```

```python
df['new_text'] = new_text
```

```python
for i,j in zip(df['text'],df['new_text']):
    print("Original")
    print(i)
    print('\n' + 'New')
    print(j + '\n')
```
