```python
from pydub import AudioSegment
from pydub.generators import WhiteNoise
import pandas as pd
from pathlib import Path
import numpy as np
```

# White Noise

```python
file = pd.read_csv('data/ner_pred/ner_pred_con2.csv', usecols=['start_time', 'end_time'])
```

```python
def white_noise(file, input_path, output_path):
    file = pd.read_csv(file, usecols = ['start_time', 'end_time']) #ไฟล์ csv ที่ต้องตัด word ออก พร้อมระบุเวลา
    sound = AudioSegment.from_file(input_path) #ไฟล์เสียงต้นฉบับ
    begin = 0
    count_check = 0
    new_sound = sound[0:0]
    for start, stop in zip(file['start_time'], file['end_time']):
        normal = sound[begin * 1000:start * 1000] # * 1000 เพื่อให้ทำให้เป็น millisec
        begin = stop
        white_n = WhiteNoise().to_audio_segment(duration = ((stop*1000) - (start * 1000))).apply_gain(-35)
        new_sound += normal + white_n
        count_check += 1
        if count_check == len(file):
            new_sound += sound[begin * 1000:]
            new_sound.export(out_f = output_path, format = "wav")
    
    return new_sound
```

```python
new_sound2 = white_noise("data/ner_pred/ner_pred_con2.csv", "data/Voice files/conversation_2.wav", "Output/conversation_2.wav")
new_sound2
```
