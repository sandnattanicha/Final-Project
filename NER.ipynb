```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import pandas as pd
from collections import Counter
```

```python
text = pd.read_csv('data/text.csv')
```

```python
text = text[['text']].dropna()
text
```

```python
def encode(text):
    word2idx = {word: idx for idx, word in enumerate(set(text.split()))}
    idx2word = {idx: word for word, idx in word2idx.items()}
    text_enc = list(map(lambda x: word2idx[x], text.split()))
    return text_enc, word2idx, idx2word
```

```python
def preprocess(sent):
    sent = nltk.word_tokenize(sent)
    sent = nltk.pos_tag(sent)
    return sent
```

```python
lst = []
for index, row in text.iterrows():
    for i in row:
        lst.append(i)
print(len(lst))
```

```python
text_code = preprocess(lst[0])
text_code
```

```python
lst_count = []
for i in lst:
    text_code = preprocess(i)
    tags = nltk.pos_tag(i)
    counts = Counter(tag for word,  tag in tags)
    lst_count.append(counts)
#     print(text_code, '\n')
```

```python
tmp = list(map(lambda x: dict(x), lst_count))
df = pd.DataFrame(tmp)
df = pd.concat([text, df], axis=1, join='inner')
df.drop(columns=['WRB', 'PRP', 'RB', 'NNS'])
```
