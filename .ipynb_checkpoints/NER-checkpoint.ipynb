```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import pandas as pd
```

```python
text = pd.read_csv('data/text.csv')
text
```

```python
text = text[['text']].dropna()
text
```

```python
lst = []
for index, row in text.iterrows():
    for i in row:
        lst.append(i)
```

```python
# for i in lst:
#     print(i)
lst[0]
```

```python
def encode(text):
    word2idx = {word: idx for idx, word in enumerate(set(text.split()))}
    idx2word = {idx: word for word, idx in word2idx.items()}
    text_enc = list(map(lambda x: word2idx[x], text.split()))
    return text_enc, word2idx, idx2word
```

```python
def preprocess(sent):
    sent = nltk.word_tokenize(sent)
    sent = nltk.pos_tag(sent)
    return sent
```

```python
for i in lst:
    text_code = preprocess(lst[i])
    text_code
```

```python
for i in text_code:
    if 'NNP' in i:
        text_code.remove(i)
    elif 'CD' in i:
        text_code.remove(i)
    elif 'CC' in i:
        text_code.remove(i)
```

```python
import pandas as pd
df = pd.DataFrame({'text':[i for i,j in text_code],'POS':[j for i,j in text_code]})
df
```

```python
df = df[df['POS'] != 'NNP']
df = df[df['POS'] != 'CD']
df = df[df['POS'] != 'CC']
df = df[df['POS'] != 'JJ']
```
